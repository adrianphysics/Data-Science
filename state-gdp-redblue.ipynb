```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib as mpl
import matplotlib.dates as mdates
import datetime as dt
import plotly.graph_objects as go
import time
import seaborn as sns
```

```python
#See the Data
state_population = pd.read_csv('../input/states/state_population.csv')
state_election = pd.read_csv('../input/states/state_elections.csv')
state_gdp = pd.read_csv('../input/state-gdp/state_gdp.csv')
```

```python
display(state_population.head())
display(state_election.head())
display(state_gdp.head())
```

```python
#Merge the tables
state_merge = pd.merge(state_population,state_election,on='State',how='outer')
state_merge = pd.merge(state_merge,state_gdp,on='State',how='outer')

#Rename columns
state_merge.rename(columns={'Pop':'Population','rank':'Rank','density':'Density','win':'Win','stateGDP':'GDP'}, inplace=True)
state_merge.set_index('Rank',inplace=True)

state_merge.to_csv('state_merge.csv')
```

```python
blue_states = state_merge[state_merge['Win']=='D'].copy()
red_states = state_merge[state_merge['Win']=='R'].copy()

print('Blue States')
display(blue_states)

print('Red States')
display(red_states)
```

```python
#GDP by State and politics
gdp = state_merge.groupby(['Win'])['GDP'].agg(['sum'])

gdp.rename(columns={'sum':'GDP'}, inplace=True)
display(gdp)
```

```python
plt.figure(figsize=(10,20))
sns.barplot(x=gdp.index,y='GDP',data=gdp)
plt.title('GDP Red vs Blue',fontsize=20)
plt.xlabel('Party',fontsize=20)
plt.ylabel('Total GDP',fontsize=20)
plt.show()

#fix color
```
